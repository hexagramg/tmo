# Ход работы

```python
import pandas as pd
import seaborn as sn
```

```python
dt = pd.read_csv('StudentsPerformance.csv', header=0)
```

```python
dt
```

```python
dt.shape
```

```python
dt.dtypes
```

```python
dt.describe()
```

```python
dt['gender'] = dt['gender'].astype('category')
dt['race/ethnicity'] = dt['race/ethnicity'].astype('category')
dt['parental level of education'] = dt['parental level of education'].astype('category')
dt['test preparation course'] = dt['test preparation course'].astype('category')
dt['lunch'] = dt['lunch'].astype('category')

```

```python
dt['gender_cat'] = dt['gender'].cat.codes
dt['race/ethnicity_cat'] = dt['race/ethnicity'].cat.codes
dt['parental level of education_cat'] = dt['parental level of education'].cat.codes
dt['test preparation course_cat'] = dt['test preparation course'].cat.codes
dt['lunch_cat'] = dt['lunch'].cat.codes
```

```python
dt.describe()
```

```python
dt
```

```python
dt.corr()
```

```python
sn.heatmap(data=dt.corr(), annot=True, fmt='.2f');
```

```python
sn.pairplot(dt)
```

```python
sn.jointplot(data=dt, x='parental level of education_cat', y='race/ethnicity_cat', kind='kde')
```

```python
sn.heatmap(pd.crosstab(dt['parental level of education_cat'], dt['race/ethnicity_cat']))
```

```python
sn.jointplot(data=dt, x='math score', y='race/ethnicity_cat', kind='reg')
```

```python
sn.lmplot(data=dt, x='math score', y='test preparation course_cat', hue='lunch_cat')
```
